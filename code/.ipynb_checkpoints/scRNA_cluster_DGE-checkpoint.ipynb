```python
""" IMPORT PACKAGES """
import pandas as pd
from sklearn.cluster import KMeans
import scanpy as sc
import matplotlib.pyplot as plt
import numpy as np
```

```python
""" LOAD DATA """
features = pd.read_csv("HsDrop.data.features.tsv", sep = '\t', header = None).iloc[:,1]
barcodes = pd.read_csv("HsDrop.data.barcodes.tsv.gz", sep='\t', compression = 'gzip', header = None)
mtx = sc.read("HsDrop.data.mtx",delimiter='\t',cache=True).transpose()
md = pd.read_csv('Hs.metadata.tsv', sep='\t') #includes all nuc-seq and scrna-seq data
metadata = md[md['library_preparation_protocol__ontology_label']=='Drop-seq'] #use only whole cell data 
metadata.shape # expect 28465 scRNA-seq objects
```

```python
""" CREATE ANNDATA OBJECT """
# create AnnData object 
adata = sc.AnnData(mtx)

# add observation names and variable names
adata.obs_names = barcodes
adata.var_names = features 
```

```python
""" ADD UMAP EMBEDDING """
umap_coords = pd.read_csv("Hs.umap.scp.tsv", sep = '\t', index_col = 0).loc[barcodes[0]]
adata.obsm['X_umap'] = umap_coords.values.astype(float)
umap_coordinates = adata.obsm["X_umap"]
kmeans = KMeans(n_clusters = 5, random_state=1).fit(umap_coordinates)
adata.obs['kmeans'] = kmeans.labels_ #adding labels as metadata
adata.obs['kmeans'] = adata.obs['kmeans'].astype(str)

# Pre-labeling of cell populations available with UMAP embedding, data was kmeans clustered and annotated accordingly 
kmeans = KMeans(n_clusters=5, random_state=1).fit(umap_coordinates)
adata.obs['kmeans'] = kmeans.labels_.astype(str)  # Adding labels as metadata
label_to_word = {
    '0': 'Mesothelial cell',
    '1': 'T cell',
    '2': 'ASPC',
    '3': 'Macrophage',
    '4': 'Endothelial Cell'
}

cpallette = ["#D88C9A", '#F2D0A9', '#99C1B9', '#3AAFB9', '#D282A6']

adata.obs['kmeans_word'] = adata.obs['kmeans'].map(label_to_word)
sc.settings.set_figure_params(dpi=100, dpi_save=300, format='png', frameon=True, transparent=False, fontsize=10)
sc.pl.umap(adata, color='kmeans_word', legend_fontsize=10, legend_loc='on data',  size=20, alpha=0.7, title='UMAP Plot', palette=cpallette)
```

```python
""" MACROPHAGE DATA SUBSETTING """
mp_marker_genes = ['hMac1', 'hMac2', 'hMac3'] #macrophage data was subset using these cell type labels 
adata.obs["cell_subtype"] = pd.Categorical(metadata["cell_subtype__custom"])
mp_metadata = pd.concat([metadata[metadata['cell_subtype__custom']==i] for i in mp_marker_genes])
mp_barcodes = [str(i) for i in mp_metadata["cell_id"]]
mp_subset = adata[adata.obs['cell_subtype'].isin(mp_marker_genes)]
# subset to only include macrophage scRNA-seq data
mp_umap_coords = umap_coords.loc[mp_barcodes] # get barcodes for mphages first
mp_subset.obsm['X_umap'] = mp_umap_coords.values.astype(float)
```

```python
""" PREPROCESSING """
# Data was already normalized and preprocessed, mitochondiral and ribosomal genes were removed before analysis
# mitochondrial genes, "MT-" for human, "Mt-" for mouse
mp_subset.var["mt"] = mp_subset.var_names.str.startswith("MT-")
mp_subset.var["ribo"] = mp_subset.var_names.str.startswith(("RPS", "RPL"))
genes_to_keep = ~(mp_subset.var['mt'] | mp_subset.var['ribo']) # Create a boolean mask for genes that are neither mitochondrial nor ribosomal
mp_subset = mp_subset[:, genes_to_keep] # Filter your data based on this mask

# FILTERING
sc.pp.filter_cells(mp_subset, min_genes=5000)
sc.pp.filter_genes(mp_subset, min_cells=50)
```

```python
""" FEATURE SELECTION, select top 2000 highly variable genes"""
sc.pp.highly_variable_genes(adata, n_top_genes=2000)
sc.pl.highly_variable_genes(adata)
```

```python
""" DIMENSIONALITY REDUCTION """
sc.tl.pca(adata)
sc.pl.pca_variance_ratio(adata, n_pcs=50, log=True)
```

```python
""" LOUVAIN CLUSTERING """
mp_marker_genes = ['hMac1', 'hMac2', 'hMac3']
adata.obs["cell_subtype"] = pd.Categorical(metadata["cell_subtype__custom"])
mp_metadata = pd.concat([metadata[metadata['cell_subtype__custom']==i] for i in mp_marker_genes])
mp_barcodes = [str(i) for i in mp_metadata["cell_id"]]
mp_subset = adata[adata.obs['cell_subtype'].isin(mp_marker_genes)]
mp_umap_coords = umap_coords.loc[mp_barcodes] # get barcodes for mphages first
mp_subset.obsm['X_umap'] = mp_umap_coords.values.astype(float)
mp_subset, adata
```

```python
# cluster using KMeans clustering and plot UMAP
kmeans = KMeans(n_clusters = 3, random_state=1).fit(mp_umap_coords) #expect 3 subpopulations, use k=3 
mp_subset.obs['kmeans'] = kmeans.labels_ #adding labels as metadata
mp_subset.obs['kmeans'] = mp_subset.obs['kmeans'].astype(str)
sc.pl.umap(mp_subset, color='kmeans', palette = 'tab10', size = 50, alpha=0.7, legend_loc = 'right_margin', return_fig=False)
```

```python
""" ANNOTATE CELLS """
# Finding marker genes 
sc.tl.rank_genes_groups(mp_subset, "kmeans", method="wilcoxon")
sc.pl.rank_genes_groups(mp_subset, n_genes=25, sharey=False)
```

```python
""" MAKE FIGURE FIT FOR PUBLICATION """
# Customize plotting parameters
sc.settings.set_figure_params(dpi=100, dpi_save=300, format='png', frameon=True, transparent=False, fontsize=10)

# Adjust UMAP plot parameters
sc.pl.umap(mp_subset, color='cluster_label', legend_loc='on data', legend_fontsize=8, size=20, alpha=0.7, title='UMAP Plot')

# Save the plot
plt.savefig('umap_plot.png', dpi=300)
```

```python
""" DEG ANALYSIS """
# Get the DEG results from the Scanpy analysis
results = mp_subset.uns["rank_genes_groups"]

# Extract information for the top DEGs
groups = results["names"].dtype.names  # Get group names
group_degs_list = []  # List to store DataFrames for each group

for group in groups:
    # Create a DataFrame for the DEGs in each group
    group_degs = pd.DataFrame({
        "Gene": results["names"][group],  # Gene symbols
        "Log-fold change": results["logfoldchanges"][group],  # Log-fold changes
        "Test statistic": results["scores"][group],  # Test statistics
        "P-value": results["pvals"][group],  # P-values
    })
    
    # Append the DataFrame to the list with its group label
    group_degs_list.append((group, group_degs))

# Print each DataFrame with its group label
for group, group_degs in group_degs_list:
    print(f"Group: {group}")
    print(group_degs)
    print("\n")
```

```python
import os

# Create a directory to store the Excel files if it doesn't exist
output_dir = "DEG_Excel_Files"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Export each DataFrame to an Excel file named after the group (cluster/group 1 is RTM)
for group, group_degs in group_degs_list:
    output_file = os.path.join(output_dir, f"{group}_DEGs.xlsx")
    group_degs.to_excel(output_file, index=False)
    print(f"Exported DEGs for group {group} to {output_file}")

#DEG were used for pathway enrichment analysis 
```
